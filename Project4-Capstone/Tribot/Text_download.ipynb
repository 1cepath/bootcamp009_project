```python
import requests 
from bs4 import BeautifulSoup

url = 'http://www.gutenberg.org/cache/epub/2349/pg2349.txt'
bible = ''
chapter_list = range(4)

for i in chapter_list:
    url = 'https://www.biblegateway.com/passage/?search=Colossians+'+str(i+1)+'&version=NIV&interface=print'

    r = requests.get(url)
    text = r.text
    soup = BeautifulSoup(text,"html")

    for link in soup.findAll("div", { "class" : "passage-text" }):
        chapter = link.text
        
    print('Chapter '+str(i+1))

    chapter = chapter[:chapter.find('Footnotes')].replace('New International Version (NIV)','')
    print(chapter[:chapter.find('Footnotes')])
    print("-"*50)
    
    bible = bible+chapter
# print(bible)
```

```python
fh = open('./books/chapter12.txt','wb')
content = bible.encode('ascii','ignore')
fh.write(content)
fh.close()
```

```python
# Hemingway
url = 'https://gutenberg.ca/ebooks/hemingwaye-sunalsorises/hemingwaye-sunalsorises-00-h.html'
book = ''
r = requests.get(url)
text = r.text
hemingway_text = text.split('CHAPTER')
hemingway_text_result = []
for chapter in hemingway_text:
    soup = BeautifulSoup(chapter,"html")
    hemingway_chapter = ''

    for link in soup.findAll("p"):
        paragraph = link.text
        hemingway_chapter = hemingway_chapter+paragraph
    
    hemingway_text_result.append(hemingway_chapter)

print(len(hemingway_text_result))
```

```python
print(hemingway_text_result[24])
```

```python
fh = open('./books/chapter3.txt','wb')
content = hemingway_text_result[24].encode('ascii','ignore')
fh.write(content)
fh.close()
```
