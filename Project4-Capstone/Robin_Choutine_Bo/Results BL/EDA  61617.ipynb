```python
sns.set(font_scale=1.3,style='whitegrid')
g=sns.countplot(y="MARKETINGCATEGORYNAME", data=df,
                   palette="BuPu")
plt.ylabel('')
plt.title('Drug Category')
```

```python
sns.set(font_scale=1.3)
sns.set_style("whitegrid")
sns.lmplot("year", "logprice",df, hue="MARKETINGCATEGORYNAME", palette="hls",fit_reg=False)
#plt.xlim(1.94e7,)   # box plot
```

```python
ranks = df.groupby("MARKETINGCATEGORYNAME")["logprice"].median().fillna(0).sort_values()[::-1].index
sns.boxplot(y='MARKETINGCATEGORYNAME', 
            x= ('logprice'), 
            data=df, 
            palette='hls',order=ranks)    # only one 3.0 data
plt.ylabel('')

plt.title('Drug Price and Marketing Catrgory ')
```

```python
plt.figure(figsize=(1.5,4))  
sns.set_style("whitegrid")
#tips = sns.load_dataset("tips")
sns.violinplot(y=df["logprice"],color='#00FFFF')
```

```python
sns.set(font_scale=1.2,style='whitegrid')
plt.figure(figsize=(10,6))   
ax=df.groupby('DOSAGEFORMNAME')['price'].mean().sort_values(ascending=True).tail(15).plot.barh(color='blue')
ax.set_xlabel("")
ax.set_ylabel("")
ax.set_title('Mean Drug Unit Price and Dosage Form (Top15)')
```

```python
sns.set(font_scale=1.2,style='whitegrid')
plt.figure(figsize=(10,6))   
ax=df.groupby('ROUTENAME')['price'].mean().sort_values(ascending=True).tail(15).plot.barh(color='blue')
ax.set_xlabel("");ax.set_ylabel("");ax.set_title('Mean Drug Unit Price and Dosing Route (Top 15)')
```

```python
sns.set_style("whitegrid")
sns.set(font_scale=1.2,style='whitegrid')
plt.figure(figsize=(10,6))
ax=df.groupby('LABELERNAME')['price'].max().sort_values(ascending=True).tail(12).plot.barh(color='r')
ax.set_ylabel("");ax.set_title('Top 12 drug prices and the Companies')
```
