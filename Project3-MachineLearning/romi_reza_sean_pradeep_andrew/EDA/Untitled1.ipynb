```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns

%matplotlib inline

sns.set(style="ticks", palette="muted", color_codes=True)
```

```python
# load dataset
df = pd.read_csv("./cleaned_df_2017-05-27.csv", parse_dates=['timestamp'], index_col=False, low_memory=False)

df.shape
```

```python
plt.figure(figsize=(18,6), dpi=120)
df['price_doc'].rolling(window=360,min_periods=10).median().dropna().plot()
sns.tsplot(df['price_doc_log'], color='grey', alpha=0.3)
plt.plot()
```

```python
plt.figure(figsize=(18,6), dpi=120)
df.groupby(['year','month']).median()['price_doc'].plot()
```
