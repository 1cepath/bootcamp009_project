Satchel Paige is skewing the numbers, so I'm going to take him out of the dataset

```python
df = df.loc[df['name'] != 'Satchel Paige',]
```

Okay, in order to calculate overall FIP, I'll have to first create a dataframe that groups by players and takes the sums of hits, walks, hrs, ip, bf, wp, bk, and hbp. Let's do this first.

```python
career_group = df.groupby(['name'])['ip', 'er', 'k','h', 'hr', 'bb', 'ibb', 'hbp', 'wp', 'bk'].agg('sum')
```

```python
np.sum(career_group.isnull())
```

Replace missing HBP with zero.

```python
career_group['hbp'] = career_group['hbp'].fillna(0)
```

Calculate ERA, FIP, WHIP, H9, HR9, BB9, K9

```python
career_group['era'] = (career_group['er'] / career_group['ip'])*9
```

```python
#FIP = (13*HR+3*BB-2*K)/IP + 3.2
career_group['fip'] = ((13*career_group['hr'] + 3*career_group['bb'] - 2*career_group['k']) / career_group['ip']) + 3.2
```

```python
career_group['whip'] = (career_group['bb'] + career_group['h']) / career_group['ip']
```

```python
career_group['h9'] = (career_group['h'] / career_group['ip']) * 9
```

```python
career_group['hr9'] = (career_group['hr'] / career_group['ip']) * 9
```

```python
career_group['bb9'] = (career_group['bb'] / career_group['ip']) * 9
```

```python
career_group['k9'] = (career_group['k'] / career_group['ip']) * 9
```
