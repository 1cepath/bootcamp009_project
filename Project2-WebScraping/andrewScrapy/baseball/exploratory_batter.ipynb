Let's make a list. The most missing values is IBB with 2427 out of 3625. It could be that at the time, IBBs weren't being recorded. Or, if the player actually wasn't intentionally, it was left blank. I'm going to replace these with zero.

```python
batters['ibb'].fillna(0)
```

```python
np.sum(batters.isnull())
```

Cool. Let's group all columns that I think I can replace missing values with zero. They are cs and tb.

```python
batters['cs'] = batters['cs'].fillna(0)
batters['tb'] = batters['tb'].fillna(0)
```

I'm curious about some of these missing BA's and OBP's and SLG. Can we impute via data provided?

```python
pd.set_option('display.max_columns', None)
batters.loc[batters.isnull().hits,:]
```

At least when checking for why hits are missing, there is no relevant info for why they are missing or why everything is gone. I think it's possibly because Scrapy scraped hidden rows that don't actually show on the webspage, or because these players played in the 1800's and don't have all available data.

```python
batters = batters.dropna(subset=['hits']) 
```

```python
batters.loc[batters.isnull().any(axis=1),:]
```

Hmmm I'm not exactly sure what to make of the 33 missing games now, I believe they are MCAR. Leave them as NA, I will primarily be focused more on PA's per season, which we are not missing any of.

```python
batters.loc[batters.isnull().games,:]
```

Deleting OPS+ cuz I have no idea what it is and if I can't understand it, how am I going to explain it to people.

```python
batters = batters.drop('ops_plus', 1)
```

There are still 5 columns in which we have over 300 missing values (so, sb, r, rb, bb). Let's look at these

```python
batters.loc[batters.isnull().ba,:]
```

Remove missing BAs, very little data available in those observations

```python
batters = batters.dropna(subset=['ba']) 
```

```python
batters.loc[batters.isnull().games,:]
```

Removing missing slg's and ops's, only 7 rows and missing a bunch of other stats

```python
batters = batters.dropna(subset=['slg']) 
```

```python
batters.loc[batters.isnull().team,:]
```

```python
batters = batters.dropna(subset=['team']) 
```

```python
batters.loc[batters.isnull().any(axis=1),:]
```

Before converting, fill NAs with a invalid number (I can't think of how else to do this besides disregarding missing data). Imputing random values seems inaccurate, and most of the counting stats are random and independent from their rates. I'll fill them in with an invalid number (-999), that way I know that they were missing. 

```python
batters['bb'] = batters['bb'].fillna(-999)
```

```python
batters['bb'] = batters['bb'].astype(int)
batters['pa'] = batters['pa'].astype(int)
batters['bb_rate'] = batters['bb'] / batters['pa']
```

```python
# Now so rate
batters['so'] = batters['so'].fillna(-999)
```

```python
batters['so'] = batters['so'].astype(int)
```

```python
batters['so_rate'] = batters['so'] / batters['pa']
```

```python
#drop tb
batters = batters.drop('tb', 1)
```
