Covert "None" to NaN.

```python
pitchers = pitchers.replace('None', np.nan)
```

Okay. Where do we start. Let's just take a look at some of these rows. Rember, we don't want to get rid of too many, we only have 1600 rows.

```python
pd.set_option('display.max_columns', None)
pitchers.loc[pitchers['ibb'].isnull(),:]
```

Replacing all missing IBB's with 0

```python
pitchers['ibb'] = pitchers['ibb'].fillna(0)
```

Because there are a few relievers in the HOF, I'm going to fill any missing saves with 0, as well as any CG or SHO

```python
pitchers.loc[pitchers['k'].isnull(),:]
```

```python
pitchers['cg'] = pitchers['cg'].fillna(0)
pitchers['sho'] = pitchers['sho'].fillna(0)
pitchers['saves'] = pitchers['saves'].fillna(0)
```

```python
pitchers.loc[pitchers['r'].isnull(),:]
```

Remove anything without "run"

```python
pitchers = pitchers.dropna(subset=['r']) 
```

```python
pitchers.loc[pitchers['name'] == "John Smoltz",:]
```

```python
pitchers.loc[pitchers['fip'].isnull(),:]
```

Of what we have, I'm hesitant to get rid of these counting stats. I'll leave them blank for now. They could be important if I want to manually calculate FIP (if possible) or ERA. I also don't want to delete anymore rows.

Worth noting, if we decide to take the average counts/rates per pitcher, we might want to not include the relievers who are in the HOF.
