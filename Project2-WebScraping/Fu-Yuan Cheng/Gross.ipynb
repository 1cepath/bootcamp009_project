```python
data_S_ROI=data[data.ROI<100]
data_S_ROI = data_S_ROI.reset_index(drop=True)
```

```python
plt.figure(figsize=(13,6))
invYr = data_S_ROI.groupby('year', as_index=False).gross.mean()
sns.barplot(x='year', y='gross', data=invYr)
plt.savefig('year_gross')
```

```python
plt.figure(figsize=(13,6))
invYr = data_S_ROI.groupby('year', as_index=False).budget.mean()
sns.barplot(x='year', y='budget', data=invYr)
plt.savefig('year_budget')
```

```python
plt.figure(figsize=(13,6))
invYr = data_S_ROI.groupby('year', as_index=False).ROI.mean()
sns.barplot(x='year', y='ROI', data=invYr)
plt.savefig('year_ROI')
```
