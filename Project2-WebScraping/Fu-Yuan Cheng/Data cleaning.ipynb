# Split rating & audinece

```python
def split_f(x):
    return x.split(' ')
rate_temp=data_1.rate.apply(split_f)
```

```python
x=rate_temp.apply(len)
x.unique()
```

```python
def extract_f(x):
    return x[0]
data_1.rating=rate_temp.apply(extract_f)
```

```python
def rate_audi(x):
    temp = x['rate'].split(' ')
    x['rating']=temp[0]
    x['audience']= temp[3]
    return x
data_1=data_1.apply(rate_audi, axis=1 )
```

```python
x=data_1.audience
L=[]
for i in x:
    L.append(int(re.sub(",","",i)))
data_1.audience=pd.Series(L)
```

```python
x=data_1.rating
L=[]
for i in x:
    L.append(float(re.sub(",","",i)))
data_1.rating=pd.Series(L)
```

# Clean length

```python
data_1.length=data_1.length.apply(lambda x: x.split('T')).apply(lambda x: x[1])
```

```python
data_1.length=data_1.length.apply(lambda x: x.split('M')).apply(lambda x: x[0])
```

```python
data_1.length=data_1.length.apply(int)
```

# Gross Profit

```python
x=data_1.gross.apply(lambda x: x.strip())
x=x.apply(lambda x: x[1:])
```

```python
L=[]
for i in x:
    L.append(int(re.sub(",","",i)))
data_1.gross=pd.Series(L)
```

# Budget

```python
data_1.budget=data_1.budget.apply(lambda x: x.strip())
set(data_1.budget.apply(lambda x: x[0]))
data_1=data_1[data_1.budget.str.startswith("$")]
data_1 = data_1.reset_index(drop=True)
```

```python
x=data_1.budget.apply(lambda x: x[1:])
L=[]
for i in x:
    L.append(int(re.sub(",","",i)))
data_1.budget=pd.Series(L)
```

# Open Week

```python
x=data_1.open_week
x=data_1.open_week.str.replace("\n","")
x=x.apply(lambda x: x.strip())
x=x.apply(lambda x: x.split("       "))
y=x.apply(lambda x: x[0])
data_1.open_week=y
data_1=data_1[y.str.startswith("$")]
data_1 = data_1.reset_index(drop=True)
```

```python
x=data_1.open_week.apply(lambda x: x[1:])
L=[]
for i in x:
    L.append(int(re.sub(",","",i)))
data_1.open_week=pd.Series(L)
```
