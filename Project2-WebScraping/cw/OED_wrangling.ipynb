```python
import pandas as pd
import numpy as np
import re
```

```python
oed = pd.read_csv('oed.csv')
oed.head()
```

```python
oed['entry_type']=oed['entry_type'].map(str)
oed['word']=oed['word'].map(str)
oed['month']=oed['month'].map(str)
oed['year']=oed['year'].map(str)
oed['part_of_speech']=oed['part_of_speech'].map(str)
```

```python
oed['acronym'] = oed['word'].map(lambda x: x==x.upper())
oed[oed['acronym']==True]
```

```python
oed['dash'] =oed['word'].map(lambda x: re.search('-', x.lower()) != None)
oed['dash'].value_counts()

np.unique(oed[oed['dash']==True]['word'])
```

```python
oed['word_length'] = oed['word'].map(lambda x: len(str(x).split(' ')))
oed.head()
```

```python
oed['word_length'].value_counts()
```

```python
oed.sort_values('word_length', ascending=False)
```

```python
ud = pd.read_csv('ud_2.csv')
ud = ud.dropna()
ud['phrase']=ud['phrase'].map(str)

oed['word']=oed['word'].map(lambda x: x.lower())
ud['phrase']=ud['phrase'].map(lambda x: x.lower())
ud.head()
```

```python
common_words = set(oed['word'].map(lambda x: x.lower())).intersection(ud['phrase'].map(lambda x: x.lower()))
```

```python
len(common_words)
```

```python
common_words
```

```python
ud_common = ud[ud['phrase'].isin(list(common_words))]
```

```python
oed_common = oed[oed['word'].isin(list(common_words))]
```

```python
pd.merge(oed_common, ud_common, how='outer', left_on='word', right_on='phrase')
```
